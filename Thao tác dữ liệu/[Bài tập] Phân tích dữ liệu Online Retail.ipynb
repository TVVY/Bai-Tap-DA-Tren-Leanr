```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/OnlineRetail.csv',header=0 , encoding = "ISO-8859-1")
data
```

```python
# Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.dtypes
```

```python
print(f'Số Hàng trong dữ liệu là {data.shape[0]}, Số cột là {data.shape[1]}')
```

```python
# Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3;
# nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; 
# nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  
# nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;
data['Month'] = pd.to_datetime(data['InvoiceDate']).dt.month
conditions1 = [(data.loc[:,'Month'] == 1) | (data.loc[:,'Month'] == 2) | (data.loc[:,'Month'] == 3),
              (data.loc[:,'Month'] == 4) | (data.loc[:,'Month'] == 5) | (data.loc[:,'Month'] == 6),
              (data.loc[:,'Month'] == 7) | (data.loc[:,'Month'] == 8) | (data.loc[:,'Month'] == 9)]
values1 = ['1','2','3']
data['Previous']=np.select(conditions1, values1, default='4')
data
```

```python
# Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice
data['Amount'] = data.loc[:,'Quantity'] * data.loc[:,'UnitPrice']
```

```python
data
```

```python
# Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.
new = [(data.loc[:,'Country'] == 'United Kingdom') & (data.loc[:,'Previous'] == '4'),
       (data.loc[:,'Country'] == 'France')]
new1 = ['10','5']
data['Discount(%)']=np.select(new, new1, default='0')
data
```

```python
data['Discount(%)'] = data['Discount(%)'].astype('int64')
```

```python
# Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.
data['Total'] = data['Amount'] - data['Discount(%)']
data
```
