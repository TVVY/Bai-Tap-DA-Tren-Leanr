```python
import pandas as pd
import numpy as np
data = pd.read_excel('C:/Users/Pyter/DA/BaiTap/house_price_dống-da.xlsx',header=0)
data.head()
```

```python
data.info()
print(f'Số dòng trong dữ liệu là {data.shape[0]}, Số cột là {data.shape[1]}')
```

```python
# Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên

data[(data['type_of_land']=='Bán nhà riêng\n ') & (data['ward_name']=='Phường Trung Liệt') |
     (data['type_of_land']=='Bán nhà riêng\n ') & (data['ward_name']=='Phường Khâm Thiên')] 
```

```python
# Lọc các thông tin Địa chỉ, Giá, Hướng nhà, 
# Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.
data[(data['land_certificate'] == 'Sổ đỏ') & (data['bedroom'] >= 3 )]\
.filter(['address','price','house_direction','balcony_direction','land_certificate','bedroom'])
```

```python
# Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.
data.pivot_table(values= 'price', index= 'type_of_land', aggfunc= {max,min,np.mean})
```

```python
# Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường.
data.pivot_table(values= ['bedroom','toilet','floor'], index= 'ward_name', aggfunc= np.mean)
```
