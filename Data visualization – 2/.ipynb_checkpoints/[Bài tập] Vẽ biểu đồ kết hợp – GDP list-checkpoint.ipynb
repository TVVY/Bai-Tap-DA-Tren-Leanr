```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/GDPlist.csv',header=0 , encoding = "ISO-8859-1")
data
```

Câu 1:\
Biểu đồ để hiển thị giá trị cụ thể và so sánh GDP các nước Vietnam,  Indonesia, Cambodia, Thailand và Malaysia.

```python
df = data[(data['Country']==' Vietnam')|(data['Country']==' Indonesia')|(data['Country']==' Cambodia')|(data['Country']==' Thailand')|(data['Country']==' Malaysia')]
df
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(6, 5)
ax.bar(df['Country'],df['GDP (millions of US$)'])
for bars in ax.containers:
    ax.bar_label(bars)
```

Câu 2:\
Biểu đồ để đánh giá tỉ lệ đóng góp GDP của các nước trên tổng số GDP của 5 nước Vietnam,  Indonesia, Cambodia, Thailand và Malaysia.

```python
hinh1 = ['tổng số GDP của 5 nước','GDP của Việt Nam']
sizes1 = [df['GDP (millions of US$)'].sum(),df[df['Country']==' Vietnam']['GDP (millions of US$)']]

hinh2 = ['tổng số GDP của 5 nước','GDP của Indonesia']
sizes2 = [df['GDP (millions of US$)'].sum(),df[df['Country']==' Indonesia']['GDP (millions of US$)']]

hinh3 = ['tổng số GDP của 5 nước','GDP của Cambodia']
sizes3 = [df['GDP (millions of US$)'].sum(),df[df['Country']==' Cambodia']['GDP (millions of US$)']]

hinh4 = ['tổng số GDP của 5 nước','GDP của Thailand']
sizes4 = [df['GDP (millions of US$)'].sum(),df[df['Country']==' Thailand']['GDP (millions of US$)']]

hinh5 = ['tổng số GDP của 5 nước','GDP của Malaysia']
sizes5 = [df['GDP (millions of US$)'].sum(),df[df['Country']==' Malaysia']['GDP (millions of US$)']]

hinh6 = ['Cambodia','Indonesia','Malaysia','Thailand','Vietnam']
sizes6 = [df[df['Country']==' Cambodia']['GDP (millions of US$)'],df[df['Country']==' Indonesia']['GDP (millions of US$)'],\
         df[df['Country']==' Malaysia']['GDP (millions of US$)'],df[df['Country']==' Thailand']['GDP (millions of US$)'],\
         df[df['Country']==' Vietnam']['GDP (millions of US$)'],]
```

```python

fig, ax = plt.subplots(3, 2)
fig.set_size_inches(20, 20)
ax[0][0].pie(sizes1,labels=hinh1, autopct='%1.2f%%', pctdistance=0.7, shadow=True)
ax[0][1].pie(sizes2,labels=hinh2, autopct='%1.2f%%', pctdistance=0.7, shadow=True)
ax[1][0].pie(sizes3,labels=hinh3, autopct='%1.2f%%', pctdistance=0.7, shadow=True)
ax[1][1].pie(sizes4,labels=hinh4, autopct='%1.2f%%', pctdistance=0.7, shadow=True)
ax[2][0].pie(sizes5,labels=hinh5, autopct='%1.2f%%', pctdistance=0.7, shadow=True)

ax[0][0].legend(fontsize = 8, loc = 'upper left')
ax[0][1].legend(fontsize = 8, loc = 'upper right')
ax[1][0].legend(fontsize = 8, loc = 'upper left')
ax[1][1].legend(fontsize = 8, loc = 'upper left')
ax[2][0].legend(fontsize = 8, loc = 'upper left')

plt.show()
```
