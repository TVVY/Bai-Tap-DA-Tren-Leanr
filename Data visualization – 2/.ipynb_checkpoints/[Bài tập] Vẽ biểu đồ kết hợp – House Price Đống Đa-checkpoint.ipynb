```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/house_price_Dống-Da_Hà-Nội_subdata.csv',header=0)
data
```

Câu 1:\
Phân tích mối liên hệ giữa diện tích với giá nhà. Đồng thời, giữa số phòng ngủ với giá nhà và giữa số toilet với giá nhà.


```python
fig, axs = plt.subplots(1,3)
fig.set_size_inches(20, 6)
axs[0].scatter(data['area'], data['price'], s=100, alpha=0.5)
axs[0].set_title('Mỗi liên hệ giữa số area với giá nhà')
axs[0].set_xlabel('area', fontsize = 12)
axs[0].set_ylabel('Giá nhà', fontsize = 12)

axs[1].scatter(data['bedroom'], data['price'], s=100, alpha=0.5)
axs[1].set_title('Mỗi liên hệ giữa số bedroom với giá nhà')
axs[1].set_xlabel('bedroom', fontsize = 12)
axs[1].set_ylabel('Giá nhà', fontsize = 12)

axs[2].set_title('Mỗi liên hệ giữa số toilet với giá nhà')
axs[2].scatter(data['toilet'], data['price'], s=100, alpha=0.5)
axs[2].set_xlabel('toilet', fontsize = 12)
axs[2].set_ylabel('Giá nhà', fontsize = 12)

plt.show()
```

Câu 2:\
So sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land). Đồng thời thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land).

```python
data.dropna(subset=['area'])
data["giá (triệu/m2)"] = data.loc[:,'price'] / data.loc[:,'area']
```

```python
fig, axs = plt.subplots(1,2)
fig.set_size_inches(20, 6)
axs[0].bar(data['property_type'], data["giá (triệu/m2)"])
axs[0].set_title('Biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land)', fontsize = 14)
axs[0].set_xlabel('Dạng nhà', fontsize = 12)
axs[0].set_ylabel('Giá (triệu/m2)', fontsize = 12)

labels = data['property_type'].unique()
sizes = data['property_type'].value_counts()
axs[1].pie(sizes, labels = labels, autopct='%1.2f%%', pctdistance=0.7, shadow=True)
axs[1].set_title('Biểu đồ thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land)', fontsize = 14)

plt.show()
```

Câu 3:\
Vẽ biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ, theo số phòng toilet hoặc theo diện tích.

```python
fig, axs = plt.subplots(1,3)
fig.set_size_inches(22, 6)

axs[0].bar(data['bedroom'], data["giá (triệu/m2)"])
axs[0].set_title('Biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ', fontsize = 10)
axs[0].set_xlabel('Số lượng phòng ngủ', fontsize = 12)
axs[0].set_ylabel('Giá (triệu/m2)', fontsize = 12)


axs[1].bar(data['toilet'], data["giá (triệu/m2)"])
axs[1].set_title('Biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng toilet', fontsize = 10)
axs[1].set_xlabel('Số lượng phòng toilet', fontsize = 12)
axs[1].set_ylabel('Giá (triệu/m2)', fontsize = 12)

axs[2].bar(data['area'], data["giá (triệu/m2)"])
axs[2].set_title('Biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo diện tích', fontsize = 10)
axs[2].set_xlabel('diện tích', fontsize = 12)
axs[2].set_ylabel('Giá (triệu/m2)', fontsize = 12)

plt.show() 
```
