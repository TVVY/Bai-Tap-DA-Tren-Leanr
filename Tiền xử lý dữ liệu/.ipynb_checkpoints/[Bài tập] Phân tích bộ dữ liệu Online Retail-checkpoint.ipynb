```python
print('Các cột chứa dữ liệu khuyết thiếu là:')
for index, column in enumerate(data.columns):
    temp = data.loc[:,column].isnull()
    if temp.sum() > 0:
        print('cột ' , data.columns[index], ' chứa ' , temp.sum() , ' giá trị KT')
print('Tổng số lượng dòng có chứa giá trị khuyết thiếu là: ', data.shape[0] - data.dropna().shape[0])
print("Các cột 'Description  ', 'CustomerID  ' khổng thể xóa")
print("Bởi vì số lượng khuyết thiếu là nhiều, và nếu thiếu các dữ liệu này thì dòng đấy sẽ mất đi nhiều ý nghĩa và thông tin của khách hàng.")
```

```python
data.fillna(value= {'Description':'Không biết'}, inplace=True)
```

```python
data[(data['Description']=='Không biết')]
```

```python
df = data[['Quantity','UnitPrice']]
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
data1 = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f'Số giá trị ngoại lai bị bỏ đi là {df.shape[0]-data1.shape[0]}, \
Số dữ liệu còn lại là {data1.shape[0]}')
```
