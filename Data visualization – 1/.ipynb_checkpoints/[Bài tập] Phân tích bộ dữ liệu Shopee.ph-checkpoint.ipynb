```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/shopeep_koreantop_clothing_shop_data.csv',header=0)
data
```

CÃ¢u 1:\
Váº½ biá»ƒu Ä‘á»“ so sÃ¡nh sá»‘ lÆ°á»£ng shop gia nháº­p theo cÃ¡c nÄƒm.

```python
plt.bar(data['join_year'].unique(), data['join_year'].value_counts())
plt.title('Biá»ƒu Ä‘á»“ so sÃ¡nh sá»‘ lÆ°á»£ng shop gia nháº­p theo cÃ¡c nÄƒm', fontsize = 14)
plt.xlabel('NÄƒm', fontsize = 12)
plt.ylabel('Sá»‘ lÆ°á»£ng ', fontsize = 12)
plt.show() 
```

CÃ¢u 2:\
Váº½ biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n má»‘i quan há»‡ giá»¯a tá»‰ lá»‡ pháº£n há»“i vá»›i sá»‘ lÆ°á»£t khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ tá»‘t.

```python
plt.scatter(data['response_rate'], data['rating_good'])
plt.title('Biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n má»‘i quan há»‡ giá»¯a tá»‰ lá»‡ pháº£n há»“i vá»›i sá»‘ lÆ°á»£t khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ tá»‘t', fontsize = 14)
plt.ylabel('khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ tá»‘t', fontsize = 12)
plt.xlabel('tá»‰ lá»‡ pháº£n há»“i', fontsize = 12)
plt.show()
```

CÃ¢u 3:\
Váº½ biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n má»‘i quan há»‡ giá»¯a thá»i gian pháº£n há»“i (Ä‘Æ¡n vá»‹ giÃ¢y) vá»›i sá»‘ lÆ°á»£t khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ xáº¥u.

```python
data.loc[:,'Hour']=pd.to_datetime(data.loc[:,'response_time'],format=' %H:%M:%S').dt.hour
data.loc[:,'Minute']=pd.to_datetime(data.loc[:,'response_time'],format=' %H:%M:%S').dt.minute
data.loc[:,'Second']=pd.to_datetime(data.loc[:,'response_time'],format=' %H:%M:%S').dt.second
data['GiÃ¢y'] = (data['Hour']*3600) + (data['Minute']*60) + data['Second']
data.loc[:,['response_time','Hour', 'Minute', 'Second','GiÃ¢y']]
```

```python
plt.scatter( data['GiÃ¢y'],data['rating_bad'])
plt.title('Biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n má»‘i quan há»‡ giá»¯a thá»i gian pháº£n há»“i (Ä‘Æ¡n vá»‹ giÃ¢y) vá»›i sá»‘ lÆ°á»£t khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ xáº¥u', fontsize = 14)
plt.ylabel('khÃ¡ch hÃ ng Ä‘Ã¡nh giÃ¡ xáº¥u', fontsize = 12)
plt.xlabel('thá»i gian pháº£n há»“i (Ä‘Æ¡n vá»‹ giÃ¢y)', fontsize = 12)
plt.show()
```

CÃ¢u 4:\
Váº½ biá»ƒu Ä‘á»“ thá»ƒ hiá»‡n xu hÆ°á»›ng cá»§a sá»‘ lÆ°á»£ng shop gia nháº­p theo thá»i gian.

```python
df = data.groupby('join_year')['join_month'].count()
plt.plot(df.index,df, label = 'Month', linewidth = 2, c = 'r', marker = 'o')

plt.title('Biá»ƒu Ä‘á»“ so sÃ¡nh sá»‘ lÆ°á»£ng shop gia nháº­p theo thá»i gian', fontsize = 14)
plt.ylabel('sá»‘ lÆ°á»£ng shop gia nháº­p', fontsize = 12)
plt.xlabel('NÄƒm', fontsize = 12)
plt.show() 
```
