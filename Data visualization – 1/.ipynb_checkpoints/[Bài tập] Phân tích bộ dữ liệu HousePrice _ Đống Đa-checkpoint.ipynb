```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/house_price_Dống-Da_Hà-Nội_subdata.csv',header=0)
data
```

Câu 1:\
Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, giữa số phòng ngủ với giá nhà, giữa số toilet với giá nhà.

```python
plt.scatter(data['area'], data['price'], s=100, alpha=0.5)
plt.title('Mỗi liên hệ giữa số area với giá nhà')
plt.xlabel('area', fontsize = 12)
plt.ylabel('Giá nhà', fontsize = 12)
plt.show()
```

```python
plt.scatter(data['bedroom'], data['price'], s=100, alpha=0.5)
plt.title('Mỗi liên hệ giữa số bedroom với giá nhà')
plt.xlabel('bedroom', fontsize = 12)
plt.ylabel('Giá nhà', fontsize = 12)
plt.show()
```

```python
plt.scatter(data['toilet'], data['price'], s=100, alpha=0.5)
plt.title('Mỗi liên hệ giữa số toilet với giá nhà')
plt.xlabel('toilet', fontsize = 12)
plt.ylabel('Giá nhà', fontsize = 12)
plt.show()
```

Câu 2:\
 Vẽ biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land).

```python
data["giá (triệu/m2)"] = data.loc[:,'price'] / data.loc[:,'area']
```

```python
plt.bar(data['property_type'], data["giá (triệu/m2)"])
fig = plt.gcf()
fig.set_size_inches(15, 6.5)
plt.title('Biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land)', fontsize = 14)
plt.xlabel('Dạng nhà', fontsize = 12)
plt.ylabel('Giá (triệu/m2)', fontsize = 12)
plt.show() 
```

Câu 3:\
Vẽ biểu đồ thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land).

```python
df = data[data['property_type']=='trong ngo'][['property_type','postDate']]
df1 = data[data['property_type']=='mat pho'][['property_type','postDate']]
```

```python
labels = df['property_type'].unique(),df1['property_type'].unique()
sizes = df['postDate'].count(),df1['postDate'].count()
plt.pie(sizes, labels = labels, autopct='%1.2f%%', pctdistance=0.7, shadow=True)
plt.title('Biểu đồ thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land)', fontsize = 14)
plt.show()
```

Câu 4:\
Vẽ biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ.

```python
plt.bar(data['bedroom'], data["giá (triệu/m2)"])
plt.title('Biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ', fontsize = 14)
plt.xlabel('Số lượng phòng ngủ', fontsize = 12)
plt.ylabel('Giá (triệu/m2)', fontsize = 12)
plt.show() 
```
