```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/Credit_Scoring.csv', encoding = "ISO-8859-1")
data
```

```python
del data['Unnamed: 0']
```

```python
for index, column in enumerate(data.columns):
    temp = data.loc[:,column].isna()
    if temp.sum() > 0:
        print('Cột', data.columns[index], 'chứa' , temp.sum() , 'giá trị KT')
```

```python
data.dropna(subset=['MonthlyIncome'],inplace=True)
```

```python
phu_thuoc = data[(data['NumberOfDependents']==0)]['MonthlyIncome']
k_phu_thuoc = data[(data['NumberOfDependents']>=1)]['MonthlyIncome']
```

# Có phải những khách hàng không có người phụ thuộc sẽ có thu nhập trung bình theo tháng (MonthlyIncome) nhỏ hơn những khách hàng có người phụ thuộc không (với mức ý nghĩa 10%)

```python
from scipy.stats import ttest_ind
# Thực hiện kiểm định t-test
alpha = 0.1  # Mức ý nghĩa 10%

t_statistic, p_value = ttest_ind(phu_thuoc, k_phu_thuoc)

if p_value < alpha:
    print("Có bằng chứng để bác bỏ giả thuyết H0.")
    if t_statistic < 0:
        print("Khách hàng không có người phụ thuộc có thu nhập trung bình thấp hơn.")
    else:
        print("Khách hàng không có người phụ thuộc có thu nhập trung bình cao hơn.")
else:
    print("Không đủ bằng chứng để bác bỏ giả thuyết H0.")
```

# Có phải trung bình số lượng khoản vay (NumberOfOpenCreditLinesAndLoans) những khách hàng gặp khó khăn trong vòng 2 năm trở lại đây (SeriousDlqin2yrs =1) thì sẽ cao hơn những khách hàng không gặp khó khăn không với mức ý nghĩa 10%

```python
df_trouble = data[data['SeriousDlqin2yrs'] == 1]
df_no_trouble = data[data['SeriousDlqin2yrs'] == 0]
```

```python

t_stat, p_value = ttest_ind(df_trouble['NumberOfOpenCreditLinesAndLoans'], df_no_trouble['NumberOfOpenCreditLinesAndLoans'], equal_var=False)

alpha = 0.1
if p_value < alpha:
    print("Có sự khác biệt về trung bình số lượng khoản vay giữa hai nhóm khách hàng.")
    if t_statistic < 0:
        print("Khách hàng gặp khó khăn có số lượng khoản vay thấp hơn.")
    else:
        print("Khách hàng gặp khó khăn có số lượng khoản vay cao hơn.")
else:
    print("Không đủ bằng chứng để kết luận rằng có sự khác biệt về trung bình số lượng khoản vay giữa hai nhóm khách hàng.")
```
