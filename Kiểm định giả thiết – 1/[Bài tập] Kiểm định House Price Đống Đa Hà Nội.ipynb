```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/house_price_Dống-Da_Hà-Nội_subdata.csv',header=0)
data
```

# Vẽ biểu đồ so sánh phân phối giá (triệu đ/m2) giữa nhà Phố và Nhà ngõ

```python
data['giá (triệu/m2)'] = data['price'] / data['area']
data.dropna(subset = ['giá (triệu/m2)'], inplace = True)
data
```

```python
sns.kdeplot(data['giá (triệu/m2)'])
plt.show()
```

# Kiểm định giả thuyết giá (triệu đ/m2) nhà mặt phố cao hơn giá nhà trong ngõ với mức ý nghĩa 5%

```python
from scipy import stats

trong_ngo = data[data['property_type']=='trong ngo']['giá (triệu/m2)']
mat_pho = data[data['property_type']=='mat pho']['giá (triệu/m2)']

t_statistic, p_value = stats.ttest_ind(mat_pho, trong_ngo, equal_var=True)

alpha = 0.05

if p_value < alpha:
    print("Giá nhà mặt phố cao hơn giá nhà trong ngõ với mức ý nghĩa 5%")
else:
    print("Không có bằng chứng để khẳng định giá nhà mặt phố cao hơn giá nhà trong ngõ với mức ý nghĩa 5%")
```

# Giá của những căn nhà không có thông tin về giấy tờ pháp lý thấp hơn giá nhà những căn có thông tin về giấy tờ pháp lý với mức ý nghĩa 5%

```python
values = {'land_certificate':'không có thông tin'}
data.fillna(value=values, inplace=True)
```

```python
so_do = data[data['land_certificate']=='So do']['giá (triệu/m2)'].to_frame()
kctt = data[data['land_certificate']=='không có thông tin']['giá (triệu/m2)'].to_frame()

t_stat, p_value = stats.ttest_ind(so_do, kctt, equal_var=False, alternative='less')

alpha = 0.05

if p_value < alpha:
    print("Giá của những căn nhà không có thông tin về giấy tờ pháp lý thấp hơn giá nhà những căn có thông tin về giấy tờ pháp lý với mức ý nghĩa 5%")
else:
    print("Không có bằng chứng để khẳng định giá nhà thông tin về giấy tờ pháp lý cao hơn giá nhà không có thông tin về giấy tờ pháp lý với mức ý nghĩa 5%")
```
