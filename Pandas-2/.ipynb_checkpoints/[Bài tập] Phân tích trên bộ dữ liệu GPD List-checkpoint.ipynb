```python
import pandas as pd
import numpy as np
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/GDPlist.csv',header=0 , encoding = "ISO-8859-1")
data
```

```python
data.rename(columns = {'Continent':'Tên Châu lục'},inplace=True)
```

```python
# Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.info()
print('Số hàng là:', data.shape[0], ', Số cột là:', data.shape[1])
```

```python
# Tính giá trị lớn nhất và nhỏ nhất của GDP.
data1 = data.loc[:,'GDP (millions of US$)']
print(f'Giá trị lớn nhất của GDP là {data1.max()}')
print(f'Giá trị nhỏ nhất của GDP là {data1.min()}')
```

```python
# Hãy cho biết xu hướng phân bố dữ liệu của GDP.
data.describe()
```

```python
# Hãy cho biết châu lục nào xuất hiện nhiều nhất?
data2 = data.loc[:,'Tên Châu lục']

print(f'châu lục xuất hiện nhiều nhất là {data2.mode()[0]}')
```

```python
# Với mỗi châu lục hãy tính tổng GDP; trung bình cộng GDP.
# Hợp nhất 2 bảng này thành một bảng duy nhất gồm 3 thông tin: Tên châu lục; Tổng GDP; TBC GDP.
data3 = data.groupby('Tên Châu lục')['GDP (millions of US$)'].sum().to_frame()
data3
```

```python
data4 = data.groupby('Tên Châu lục')['GDP (millions of US$)'].mean().to_frame()
data4
```

```python
data5 = pd.merge(data3,data4,on='Tên Châu lục')
data5
```

```python
data5.columns = ['Tổng GDP','TBC GDP']
data5
```
