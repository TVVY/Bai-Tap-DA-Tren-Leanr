```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/Pyter/DA/BaiTap/house_price_Dống-Da_Hà-Nội_subdata.csv',header=0)
data
```

Yêu cầu:\
Vẽ biểu đồ xu hướng phân tích mối quan hệ giữa giá nhà với diện tích, giá nhà với số lượng phòng ngủ, nhận xét.\
Vẽ biểu đồ phân bố thể hiện phân bố của giá nhà theo các hướng, nhận xét.\
Vẽ biểu đồ tần số để đếm số nhà ở mỗi hướng nhà, nhận xét.\
Vẽ biểu đồ boxplot thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
# Phát hiện các dòng, cột chứa dữ liệu khuyết thiếu
for index, column in enumerate(data.columns):
    temp = data.loc[:,column].isna()
    if temp.sum() > 0:
        print('cột ' , data.columns[index], ' chứa ' , temp.sum() , ' giá trị KT')
print('Tổng số lượng dòng có chứa giá trị khuyết thiếu là: ', data.shape[0] - data.dropna().shape[0])
```

```python
data.dropna(subset=['area'], inplace=True)
data.dropna(subset=['price'], inplace=True)
data['house_direction'] = data['house_direction'].fillna(method='ffill')
data['house_direction'] = data['house_direction'].fillna(method='bfill')
```

```python
values = {'balcony_direction':data.loc[:,'balcony_direction'].mode()[0],
         'toilet':data.loc[:,'toilet'].mode()[0],
         'bedroom':data.loc[:,'bedroom'].mode()[0],
         'floor':data.loc[:,'floor'].mode()[0],
         'land_certificate':'không có thông tin',
          }
data.fillna(value=values, inplace=True)
data
```

```python
# Phát hiện các dòng, cột chứa dữ liệu khuyết thiếu
for index, column in enumerate(data.columns):
    temp = data.loc[:,column].isna()
    if temp.sum() > 0:
        print('cột ' , data.columns[index], ' chứa ' , temp.sum() , ' giá trị KT')
print('Tổng số lượng dòng có chứa giá trị khuyết thiếu là: ', data.shape[0] - data.dropna().shape[0])
```

```python
for i in data.loc[:,['area','bedroom']]:
    sns.lmplot(x=i, y="price", data=data)
```

Nhận Xét:\
Qua biểu chúng ta có thể thấy giữa giá nhà và diện tích tập trung ở dưới 2000 area và có xu hướng đi lên.\
Qua biểu chúng ta có thể thấy giữa giá nhà và phòng ngủ tập trung ở dưới 12 phòng ngủ và có xu hướng đi ngang.


```python
plt.scatter(y=data['price'],x = data["house_direction"])
plt.xlabel('house_direction')
plt.ylabel('price')
plt.show()
```

Nhận Xét:\
Biểu đồ thể hiện các hướng nhà tập trung ở các nhà có hướng đông, đông nam, tây bắc là nhiều. các hướng nhà này có giá cao.

```python
sns.countplot(x ='house_direction', data = data)
plt.show()
```

Nhận Xét:\
Biểu đồ thể hiện số lượng nhà có hướng đông nam, tây bắc là nhiều. các hướng nhà đang có số lượng nhà cao nhất.

```python
sns.boxplot(x='house_direction', y='price', data=data)
plt.show()
```

Nhận Xét:\
Biểu đồ thể hiện sự phân bố của các hướng nhà có hướng đông, đông nam, tây bắc là nhiều.các hướng nhà này có giá cao.
